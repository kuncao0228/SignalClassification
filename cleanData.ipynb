```python
import os
import scipy.io as sio
import shutil
BPSK_FSK_DIR = 'BPSK_or_FSK'

DataDir = "SignalFiles"
FSKPath = "Temp/FSK"
BPSKPath = "Temp/BPSK"
```

```python
for file in os.listdir(BPSK_FSK_DIR):
    filePath = os.path.join(BPSK_FSK_DIR, file)
    mat_contents = sio.loadmat(filePath)

    if mat_contents['Waveform'][0] == "FSK":
        shutil.move(filePath, FSKPath)
    else:
        shutil.move(filePath, BPSKPath)



        
    
```

```python
no_FSKFiles = len(os.listdir(FSKPath))
TrainFSKLength = (int)(.6 * no_FSKFiles)


no_BPSKPath = len(os.listdir(FSKPath))
TrainBPSKLength = (int)(.6 * no_FSKFiles)

count = 0
for file in os.listdir(FSKPath):
    filePath = os.path.join(FSKPath, file)
    TrainPath = os.path.join(DataDir, "train")
    if count < TrainFSKLength:
        shutil.move(filePath, TrainPath)
        count += 1
count = 0
for file in os.listdir(BPSKPath):
    filePath = os.path.join(BPSKPath, file)
    TrainPath = os.path.join(DataDir, "train")
    if count < TrainBPSKLength:
        shutil.move(filePath, TrainPath)
        count += 1
```

```python
no_FSKFiles = len(os.listdir(FSKPath))
TestFSKLength = (int)(.5 * no_FSKFiles)


no_BPSKPath = len(os.listdir(FSKPath))
TestBPSKLength = (int)(.5 * no_FSKFiles)

count = 0
for file in os.listdir(FSKPath):
    filePath = os.path.join(FSKPath, file)
    TestPath = os.path.join(DataDir, "test")
    ValPath = os.path.join(DataDir, "val")
    if count < TestFSKLength:
        shutil.move(filePath, TestPath)
        count += 1
    else:
        shutil.move(filePath, ValPath)
        
    
count = 0
for file in os.listdir(BPSKPath):
    filePath = os.path.join(BPSKPath, file)
    TestPath = os.path.join(DataDir, "test")
    ValPath = os.path.join(DataDir, "val")
    if count < TestBPSKLength:
        shutil.move(filePath, TestPath)
        count += 1
    else:
        shutil.move(filePath, ValPath)
```
